```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
# create canvas
width = 500
height = 500
color = (255, 127, 0)

canvas = np.full((width, height, 3), color, dtype=np.uint8)

plt.imshow(cv.cvtColor(canvas, cv.COLOR_BGR2RGB))
```

```python
text = "Hello World"
font = cv.FONT_HERSHEY_COMPLEX
ordinate = (100, 200)
color = (255, 255, 255)
font_scale = 1.5
thickness = 2

cv.putText(canvas, text, ordinate, font, font_scale, color, thickness, cv.LINE_AA)

plt.imshow(cv.cvtColor(canvas, cv.COLOR_BGR2RGB))
```

```python
# create new canvas
width, height = 700, 700
color = (204, 0, 0)

canvas2 = np.full((width, height, 3), color, dtype=np.uint8)

plt.imshow(canvas2)
```

```python
width, height = 700, 700
color = (204, 0, 0)
# create canvas
canvas3 = np.full((width, height, 3), color, dtype=np.uint8)

# write text
fonts = [
    cv.FONT_HERSHEY_SIMPLEX,
    cv.FONT_HERSHEY_COMPLEX,
    cv.FONT_HERSHEY_COMPLEX_SMALL,
    cv.FONT_HERSHEY_DUPLEX,
    cv.FONT_HERSHEY_PLAIN,
    cv.FONT_HERSHEY_SCRIPT_COMPLEX,
    cv.FONT_HERSHEY_SCRIPT_SIMPLEX
]

y_ofset = 100
color = (255, 255, 255)

for i, font in enumerate(fonts):
    text = f"Text {i+1}"
    cv.putText(canvas3, text, (100, y_ofset), font, 1, color, 2, cv.LINE_AA)
    y_ofset+= 50

plt.imshow(canvas3)

```
