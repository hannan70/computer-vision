```python
import cv2 as cv
import matplotlib.pyplot as plt 
```

```python
image = cv.imread("./assets/giraffe.png")
plt.imshow(image)
plt.axis("off");
```

```python
# convert rgb image
rgb_image = cv.cvtColor(image, cv.COLOR_BGR2RGB)
plt.imshow(rgb_image)
plt.axis("off");
```

#### Flip

```python
# 1 = Horizontal (columns, left to right)
# 0 = Vertical (rows, top to bottom)
# -1 = both

horizontal_flip = cv.flip(rgb_image, 1)
vertical_flip = cv.flip(rgb_image, 0)
both_flip = cv.flip(rgb_image, -1)

titles = ["Orginal Image", "Horizontal Image", "Vertical Image", "Both Image"]
images = [rgb_image, horizontal_flip, vertical_flip, both_flip]

plt.figure(figsize=(12, 8))

for i in range(4):
    plt.subplot(2, 2, i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis("off")

plt.tight_layout()
plt.axis("off");

```

#### Rotate

```python
rotate_clock_wise = cv.rotate(rgb_image, cv.ROTATE_90_CLOCKWISE)
rotate_counterclockwise = cv.rotate(rgb_image, cv.ROTATE_90_COUNTERCLOCKWISE)
rotate_180 = cv.rotate(rgb_image, cv.ROTATE_180)

titles = ["Original Image", "Rotate Clock Wise", "Rotate Counter Clock Wise", "Rotate 180deg"]
images = [rgb_image, rotate_clock_wise, rotate_counterclockwise, rotate_180]

for i in range(4):
    plt.subplot(2, 2, i+1)
    plt.title(titles[i])
    plt.imshow(images[i])
    plt.axis("off")

plt.tight_layout()
plt.axis("off")
```

```python
x_start, y_start = 200, 300
x_end, y_end = 400, 500

cropped_image = rgb_image[y_start:y_end, x_start:x_end]

# create plot
plt.figure(figsize=(8, 4))

plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(rgb_image)
plt.axis("off")

# plot for cropend image
plt.subplot(1, 2, 2)
plt.title("Croppend Image")
plt.imshow(cropped_image)
plt.axis("off")

plt.tight_layout()
plt.show();
```
