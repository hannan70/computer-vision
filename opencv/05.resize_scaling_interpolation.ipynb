```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
image = cv.imread("./assets/global.jpg")
```

```python
plt.imshow(image)
```

```python
# conver rgb image
rgb_image = cv.cvtColor(image, cv.COLOR_BGR2RGB)
print(rgb_image.shape)
plt.imshow(rgb_image)
```

#### Image Resizing

```python
print(rgb_image.shape)
```

```python
width, height = 500, 500
resize_image = cv.resize(rgb_image, (width, height), interpolation=cv.INTER_AREA)
plt.imshow(resize_image)
```

#### Image Scaling

```python
# for shape = (height, width, channels)
# for resize = (width, height)

scale_percentage = 50
scale_heigth = int(rgb_image.shape[0] * scale_percentage / 100)
scale_width = int(rgb_image.shape[1] * scale_percentage / 100)
```

```python
scaled_image = cv.resize(rgb_image, (scale_width, scale_heigth ), interpolation=cv.INTER_AREA)

print(f"Scaled Image Shape {scaled_image.shape}")
print(f"RGB Image Shape {rgb_image.shape}")
plt.imshow(scaled_image)
```

#### Interpolation

```python
# image size
resize_width= 600
resize_height = 600

# interpolation method
methods = {
    "Nearest": cv.INTER_NEAREST, # Copy nearest pixel, it's fast but low quality 
    "Linear": cv.INTER_LINEAR, # Average of 4 nearby pixels, Default is good
    "Area": cv.INTER_AREA, # Average is taken based on area, Good for when you want to reduce the size of the image
    "Cubic": cv.INTER_CUBIC, # Takes into account 16 pixels, Good but slow when increase the image size
    "Lanczos4": cv.INTER_LANCZOS4, # Base on many pixels , Good but slow
}

resize_images = {}

for name, method in methods.items():
    resize_images[name] = cv.resize(rgb_image, (resize_width, resize_height), interpolation=method)
    cv.imwrite(f"./assets/{name}.jpg", resize_images[name])
```

```python
# show the image
plt.figure(figsize=(12, 8))

plt.subplot(2, 3, 1)
plt.imshow(rgb_image)
plt.title("Original Image")
plt.axis("off")

for i, (name, resize_image) in enumerate(resize_images.items(), start=2):
    plt.subplot(2, 3, i)
    plt.title(name)
    plt.imshow(resize_image)
    plt.axis("off")

plt.tight_layout()
plt.show()
```
