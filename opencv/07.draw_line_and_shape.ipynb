```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
# create a canvus
width, height = 700, 700
blue = (255, 127, 0)
canvas = np.full((width, height, 3), blue, dtype=np.uint8)
rgb_canvas = cv.cvtColor(canvas, cv.COLOR_BGR2RGB) 
plt.imshow(rgb_canvas)
```

```python
cv.line(canvas, (50, 50), (650, 650), (255, 255, 255), 3)

plt.imshow(cv.cvtColor(canvas, cv.COLOR_BGR2RGB))
```

```python
cv.rectangle(canvas, (100, 100), (200, 200), (0, 255, 255), 3)
cv.rectangle(canvas, (200, 200), (400, 400), (0, 255, 255), -1)

plt.imshow(cv.cvtColor(canvas, cv.COLOR_BGR2RGB))
```

```python
cv.circle(canvas, (500, 150), 100, (0, 0, 255), 3)
plt.imshow(cv.cvtColor(canvas, cv.COLOR_BGR2RGB))
```

#### Draw an Ellipse

```python
# cv.ellipse(canvas, (start point, endpoint), (length, width), angle, start angle, end angle, color, thickness)

cv.ellipse(canvas, (300, 500),  (100, 30), 45, 0, 360, (0, 255, 0), 3)

plt.imshow(cv.cvtColor(canvas, cv.COLOR_BGR2RGB))
```

#### Draw a polygon

```python
points = np.array([ [150, 400], [380, 400], [255, 300]], np.int32)
new_points = points.reshape(-1, 1, 2)
print(new_points.shape)
print(new_points)
```

```python
cv.polylines(canvas, [new_points], True, (0, 0, 255), 3)
plt.imshow(cv.cvtColor(canvas, cv.COLOR_BGR2RGB))
```
