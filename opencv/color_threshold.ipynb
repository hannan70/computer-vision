```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("./assets/giraffe.png")
```

```python
# opencv default image is BGR
plt.imshow(image)
```

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_hsb = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
```

```python
plt.imshow(image_rgb)
```

```python
plt.imshow(image_hsb)
```

```python
lower_bound = np.array([0, 0, 190])    
upper_bound = np.array([30, 185, 255])

mask = cv2.inRange(image_hsb, lower_bound, upper_bound)

plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
plt.imshow(image_hsb)
plt.title("HSB Image")
 
plt.subplot(1, 3, 2)
plt.imshow(mask)
plt.title("Mask Image")

plt.subplot(1, 3, 3)
plt.imshow(image_rgb)
plt.title("Orginal RGB") 
plt.show()
```

```python
black_hsb = np.uint8([[[ 120, 255, 0 ]]])
black_bgr = cv2.cvtColor(black_hsb, cv2.COLOR_HSV2BGR)[0][0]
```

```python
image_result = image.copy()
image_result[mask > 0] = black_bgr
```

```python
img_res_rgb = cv2.cvtColor(image_result, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_res_rgb)
```

```python
fig, ax = plt.subplots(1, 3, figsize=(12, 6))
ax[0].imshow(image_rgb)
ax[0].set_title("Original Image")
ax[0].axis("off")

ax[1].imshow(mask)
ax[1].set_title("Mask Image")
ax[1].axis("off")

ax[2].imshow(img_res_rgb)
ax[2].set_title("output image")
ax[2].axis("off")

plt.tight_layout()
plt.show()
```
